KALKULATOR SEDERHANA

```python
print("Operasi")

print("1. Jumlah \t [+]")
print("2. Kurang \t [-]")
print("3. Kali \t [*]")
print("4. Bagi \t [/]")
print("=" *25)
```

```python
angka1 = int(input('masukkan angka: '))
angka2 = int(input('masukkan angka: '))
operasi = int(input('masukkan operasi: '))
```

```python
if operasi == 1:
  hasil = angka1+angka2
  print(f'Hasil operasi dari {angka1} + {angka2} = {hasil} ')
elif operasi == 2:
  hasil = angka1-angka2
  print(f'Hasil operasi dari {angka1} - {angka2} = {hasil} ')
elif operasi == 3:
  hasil = angka1*angka2
  print(f'Hasil operasi dari {angka1} * {angka2} = {hasil} ')
elif operasi == 4:
  hasil = angka1/angka2
  print(f'Hasil operasi dari {angka1} / {angka2} = {hasil} ')
else:
  print('Periksa lagi operasi anda.')
```

```python
while True:
  angka1 = int(input('masukkan angka: '))
  angka2 = int(input('masukkan angka: '))
  operasi = int(input('masukkan operasi: '))

  if operasi == 1:
    hasil = angka1+angka2
    print(f'Hasil operasi dari {angka1} + {angka2} = {hasil} ')
  elif operasi == 2:
    hasil = angka1-angka2
    print(f'Hasil operasi dari {angka1} - {angka2} = {hasil} ')
  elif operasi == 3:
    hasil = angka1*angka2
    print(f'Hasil operasi dari {angka1} * {angka2} = {hasil} ')
  elif operasi == 4:
    hasil = angka1/angka2
    print(f'Hasil operasi dari {angka1} / {angka2} = {hasil} ')
  elif operasi == 5:
    break
  else:
    print('Periksa lagi operasi anda.')
```
